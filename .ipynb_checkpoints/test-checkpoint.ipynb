```python
import pandas as pd
import re
import numpy as np
from easy_clean import clean #importing our module
```

```python
df = pd.read_csv(r'test_data.csv')
df #Our unclean dataset
```

```python
new = clean(df['ID']).num_summary() #new variable holds return values also
```

```python
df.loc[new['duplicates']] #retriving duplicates from new
```

```python
df.loc[new['nullvalues']] #retriving null values from new
```

```python
id1 = df['ID']
id1
```

```python
cleaned = clean(id1).clean_number(0,'median') #cleaning values setting decimal points as 0 and filling NaN values with median
```

```python
cleaned #cleaned data
```

```python
clean(cleaned).num_length() #length of each characters 
```

```python
pd.DataFrame({"ID": df['ID'],"ID cleaned": cleaned,"length": clean(cleaned).num_length()})
```
